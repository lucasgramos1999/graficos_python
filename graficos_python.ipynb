```python
import pandas as pd
```

```python
df = pd.read_csv('/content/imigrantes_canada.csv')
#Fazer a leitura do arquivo
```

```python
df
```

```python
df.info()
```

```python
#Mudar o index
df.set_index('País', inplace=True)

# A função set_index() é utilizada para definir uma coluna do DataFrame como índice.
# Neste caso, a coluna 'País' passa a ser o índice do DataFrame, substituindo o índice numérico padrão.
# O parâmetro inplace=True indica que a alteração é feita diretamente no DataFrame original,
# sem a necessidade de criar ou atribuir um novo DataFrame.
# Isso facilita operações de busca, seleção e análise de dados baseadas no nome do país.
```

```python
anos = list(map(str, range(1980, 2014)))
# Cria uma lista chamada 'anos' contendo anos de 1980 até 2013
# range(1980, 2014) gera uma sequência de números de 1980 até 2013
# map(str, ...) converte cada número inteiro em string
# list(...) transforma o resultado do map em uma lista
```

```python
brasil = df.loc['Brasil', anos]
# Cria a variável 'brasil' selecionando dados específicos do DataFrame
# df.loc[] é usado para acessar dados por rótulos (label-based indexing)
# 'Brasil' indica a linha cujo índice é o país Brasil
# 'anos' é uma lista de strings com os nomes das colunas (ex: '1980', '1981', ..., '2013
```

```python
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}

# Cria um dicionário chamado 'brasil_dict' a partir dos dados do Brasil
# 'ano': brasil.index.tolist()
#   - brasil.index contém os rótulos das colunas selecionadas (os anos)
#   - tolist() converte o índice em uma lista comum do Python
#
# 'imigrantes': brasil.values.tolist()
#   - brasil.values contém os valores numéricos associados a cada ano
#   - tolist() converte esses valores em uma lista
#
# O resultado é um dicionário estruturado, ideal para criar um novo DataFrame,
# gerar gráficos ou exportar os dados
```

```python
dados_brasil = pd.DataFrame(brasil_dict)
# Cria um DataFrame chamado 'dados_brasil' a partir do dicionário 'brasil_dict'
```

```python
dados_brasil
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
# Cria um gráfico de linha usando o Matplotlib
# plt.plot() é a função usada para desenhar gráficos de linha
# dados_brasil['ano'] define os valores do eixo X (anos)
# dados_brasil['imigrantes'] define os valores do eixo Y (quantidade de imigrantes)
```

```python
plt.figure(figsize=(8, 4))
# Define o tamanho da figura do gráfico (largura x altura
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.title('Imigração do Brasil para o Canadá de 1980 a 2013') #criação de título
plt.xlabel('Ano') #rótulo X
plt.ylabel('Número de imigrantes') #rótulo Y
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
#Cria intervalo de 5 em 5 anos no eixo "X"
plt.yticks([500, 1000, 1500, 2000, 2500, 3000])
#Cria intervalo de 500 em 500 no eixo "Y"

plt.show() #Para tirar as informações e aparecer apenas o gráfico.
#Que seja a última linha do código.
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
ax.set_title('Imigração do Brasil para o Canadá de 1980 a 2013')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
# Define que os rótulos do eixo X serão exibidos a cada 5 unidades (ex: 5 em 5 anos)
ax.set_xlabel('Ano')
ax.set_ylabel('Número de imigrantes')
plt.show()

# Cria uma figura (fig) e um eixo (ax) usando o Matplotlib
# plt.subplots() é a forma recomendada de criar gráficos
# fig representa a figura inteira (a área total do gráfico)
# ax representa o eixo onde os dados serão desenhados;
# A partir do 'ax', é possível personalizar títulos, rótulos, grades e estilos
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 5))
#Cria uma figura de 15x5 polegadas com 2 gráficos lado a lado (1 linha, 2 colunas), retornando a figura (fig) e os eixos (axs) para plotagem.

axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])
axs[0].set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Número de imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5)) # Define que os rótulos do eixo X do primeiro gráfico (axs[0]) apareçam a cada 5 unidades
axs[0].grid() #Adicionar grades ao gráfico

axs[1].boxplot(dados_brasil['imigrantes'])
# Cria um boxplot (gráfico estatístico) dos valores de imigrantes no Brasil no segundo gráfico (axs[1])
axs[1].set_title('Boxplot da imigração do Brasil para o Canadá\n1980 a 2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Número de imigrantes')
axs[1].grid()#Adicionar grades ao gráfico

plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 6)) # cria uma figura com 2 linhas e 2 colunas de subplots, definindo o tamanho da figura
fig.subplots_adjust(hspace=0.5, wspace = 0.3) # ajusta o espaçamento vertical (hspace) e horizontal (wspace) entre os subplots
fig.suptitle('Imigração dos quatro principais países da América do Sul para o Canadá (1980–2013)')

axs[0,0].plot(df.loc['Brasil', anos])
axs[0,0].set_title('Brasil')

axs[0,1].plot(df.loc['Colômbia', anos])
axs[0,1].set_title('Colômbia')

axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_title('Peru')

for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  # percorre todos os subplots e define intervalos de 5 em 5 no eixo X

for ax in axs.flat:
    ax.set_xlabel('Ano')
    ax.set_ylabel('Número de imigrantes')
    # percorre todos os subplots e define os rótulos dos eixos X (Ano) e Y (Número de imigrantes)

ymin = 0
ymax = 7000
for ax in axs.ravel():
    ax.set_ylim(ymin, ymax)
    #Eixos X e Y com valores mínimos e máximos.

plt.show()
```
